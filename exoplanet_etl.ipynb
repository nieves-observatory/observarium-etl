```python
import pandas as pd
```

```python
transits_df = pd.read_csv('raw_data/transits_nasa.csv', skiprows=range(0, 30))
del transits_df['ra']
del transits_df['dec']
del transits_df['obsname']
del transits_df['algorithm']
del transits_df['isdefault']
del transits_df['ismostprecise']
del transits_df['phase']
del transits_df['transitdepthdb']
del transits_df['ttv']
del transits_df['midpointairmass']
del transits_df['targetobsstartcalendar']
del transits_df['targetobsendcalendar']
del transits_df['fractionobservable']
transits_df.at[80,'midpointcalendar']= '08/18/2021 10:00'
transits_df.at[283,'midpointcalendar']= '08/20/2021 15:00'

transits_df.head()
```

```python
exoplanets_df = pd.read_csv('raw_data/exoplanets_nasa.csv', skiprows=range(0, 90))
pd.set_option('display.max_columns', None)
exoplanets_df.head()
```

```python
counter = 1

for i in range(len(exoplanets_df.index)):
    for j in range(len(transits_df.index)):
        if exoplanets_df.loc[i].pl_name == transits_df.loc[j].planetname:
            print(exoplanets_df.loc[i].pl_name,transits_df.loc[j].planetname)
            
print(counter)
print(len(transits_df.index))
```

```python
name, ra, dec, mag, period, duration, depth, reflink, midpoint = [] , [] , [] ,[],[] , [] ,[], [], []

for i in range(len(exoplanets_df.index)):
    for j in range(len(transits_df.index)):
        if exoplanets_df.loc[i].pl_name == transits_df.loc[j].planetname:
            name.append(transits_df.loc[j].planetname)
            ra.append(transits_df.loc[j].ra_str)
            dec.append(transits_df.loc[j].dec_str)
            mag.append(exoplanets_df.loc[i].sy_vmag)
            period.append(transits_df.loc[j].period)
            duration.append(transits_df.loc[j].transitduration)
            depth.append(transits_df.loc[j].transitdepthcalc)
            midpoint.append(transits_df.loc[j].midpointcalendar)
```

```python
complete_transits = pd.DataFrame({'name': name,
             'ra': ra,
              'dec': dec,
                'mag': mag,
                   'period': period,
                                  'duration':duration,
                                  'depth':depth,
                                  'midpoint':midpoint
                                  
             })
```
